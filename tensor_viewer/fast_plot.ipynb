```python
import scipy.misc
import numpy as np
from skimage.draw import line_aa, line
```

```python
def fast_plot(y, x, img = None):
    img_width = 1000
    img_height = 200
    
        
    if type(img) == type(None):
        img = np.ndarray((img_height, img_width, 3), dtype=np.uint8)
    img.fill(0)

    rrs = []
    ccs = []
    vals = []
    for y1,y2,x1,x2 in zip(y[:-1],y[1:], x[:-1],x[1:]):
        rr, cc, val = line_aa(y1, x1, y2, x2)
        rrs.extend(rr)
        ccs.extend(cc)
        vals.extend(val)
    img[rrs, ccs, 1] = np.array(vals)  * 255
    
    img[:,499:501, 2] = 0 
    img[:,499:501, 0] = 255 
    img[:,499:501, 1] = 0 

    return img
```

```python
x = np.arange(100)*10
y = np.array(np.array((np.sin(x)+1)/2)*200, dtype=np.int)
```

```python
fig, ax = plt.subplots(1,figsize=(20,20))
plt.imshow(fast_plot(y, x), cmap = "gray")
```
