```python
from skimage.util.shape import view_as_windows
def sliding(operator, vector, radius=50):
    """ Apply operator on a vector for a sliding window of given radius
    Args:
        operator: Callable that transforms subvectors. E.g. np.mean or np.std.
        vector: 1D numpy array
        radius: Size of sliding window
    Returns:
        Transformed vector of equal size. 
    """
    windows = view_as_windows(np.array([vector]),(1, radius))
    windows = windows.reshape(windows.shape[1], windows.shape[-1])
    out = np.empty_like(vector)
    out.fill(np.nan)
    res = operator(windows, axis=1)
    indent = len(vector) - len(res)
    out[indent:indent+len(res)] = res
    return out
```

```python
from scipy.ndimage import minimum_filter
from scipy.stats import kurtosis, skew
from scipy.interpolate import interp1d

def sarles_bimodality(vector):
    n = len(vector)
    return (skew(vector)**2+1)/(kurtosis(vector)+(3*(n-1)**2/((n-2)*(n-3))))

def fit_baseline_minfilter(vector, minradius = 1000, smooth = 100, ignore_upper_when_bimodal = True, max_slope = .001, n_splines = 25, bimodal_threshold = .6, post_smooth = 10):
    guess = gaussian_filter(minimum_filter(vector,minradius),smooth)
    
    if ignore_upper_when_bimodal and sarles_bimodality(guess) > bimodal_threshold:
        print("bimodal")
        guess[np.abs(np.gradient(guess)) >= max_slope] = None
        guess[guess>np.nanmean(guess)+np.nanstd(guess)] = None
    else:
        guess[np.abs(np.gradient(guess)) >= max_slope] = None
        if not guess[0]:
            guess[0] = vector[0]
        if not guess[-1]:
            guess[-1] = vector[-1]

    x = np.arange(len(guess))[~np.isnan(guess)]
    y = guess[~np.isnan(guess)]
    fit = gaussian_filter(interp1d(x, y, kind="linear")(np.linspace(np.min(x), np.max(x), len(guess))), post_smooth)
    return fit, guess
```

```python
from scipy.spatial import ConvexHull
def rubberband(x, y):
    # Find the convex hull
    v = ConvexHull(np.array(list(zip(x, y)))).vertices
    # Rotate convex hull vertices until they start from the lowest one
    v = np.roll(v, -v.argmin())
    # Leave only the ascending part
    v = v[:v.argmax()]

    # Create baseline using linear interpolation between vertices
    return np.interp(x, x[v], y[v])
```

```python
def remove_baseline_rubberband(vector, parabola_radius = 20, smoothing = 10):
    convex = vector+np.linspace(-parabola_radius,parabola_radius, len(vector))**2
    fit = rubberband(np.arange(len(convex)),gaussian_filter(convex, smoothing))
    corrected = convex-fit
    return corrected, fit
```

```python
def fit_minima_piecewise_linear(vector, pre_smooth =0, when_smaller_then = False):
    mins_x = minima(vector,pre_smooth)
    mins_y = vector[mins_x]
    if when_smaller_then:
        mins_x = mins_x[mins_y < when_smaller_then]
        mins_y = mins_y[mins_y < when_smaller_then]
    
    interpolated = np.interp(np.arange(len(vector)),mins_x, mins_y)
    return mins_x, mins_y, interpolated
```

```python
from scipy.optimize import minimize

def get_gaussian(centers = None, amplitudes = None, widths = None,  skewed=False):
    """ Returns a function that computes several gaussians for the given parameters 
        and returns a vector that is the sum of those gaussians
    """
    n_params = 4
    if not skew:
        n_params = 0
    if type(centers) != type(None):
        n_params -= 1
    if type(amplitudes) != type(None):
        n_params -= 1
    if type(widths) != type(None):
        n_params -= 1   
    
    def func(x, *params):
        """ The order of parameters is ctr, amp, wid, skew"""
        nonlocal centers
        nonlocal amplitudes
        nonlocal widths
                
        y = np.zeros_like(x)
        for i in range(0, len(params), n_params):#Iterate over zipped parameters for several gaussians
            rel_idx = 0
            if type(centers) != type(None):
                ctr = centers[i//n_params]
            else:
                ctr = params[i]
                rel_idx += 1
                
            if type(amplitudes) != type(None):
                amp = amplitudes[i//n_params]
            else:
                amp = params[i+rel_idx]
                rel_idx += 1
                
            if type(widths) != type(None):
                wid = widths[i//n_params]
            else:
                wid = params[i+rel_idx]
                rel_idx += 1

            if skewed:
                skew = params[i+rel_idx]
                #y = skewed_gaussian(x, ctr, amp, wid, skew)
                y = y+right_skewed_gaussian(x, amp, ctr, wid, skew)
            else:
                y = y+gaussian(x, amp, ctr, wid)#y + amp * np.exp( -((x - ctr)/wid)**2)
        return y
    return func

from scipy.stats import norm
def skewed_gaussian(x, center, height, width, skew):
    p = [height, center, width, skew]
    amp = (p[0] / (p[2] * np.sqrt(2 * np.pi)))
    spread = np.exp((-(x - p[1]) ** 2.0) / (2 * p[2] ** 2.0))
    skew = (1 + erf((p[3] * (x - p[1])) / (p[2] * np.sqrt(2))))
    skewed_gauss =  amp * spread * skew
    skewed_gauss /= np.max(skewed_gauss)
    skewed_gauss *= height
    return skewed_gauss

def right_skewed_gaussian(x, amp, ctr, wid, wid1):    
    y = gaussian(x, amp, ctr, wid)
    y1 = gaussian(x, amp, ctr, wid1)
    y[ctr:] = y1[ctr:]
    return y

def gaussian(x, amp, ctr, wid):
    return amp * np.exp( -((x - ctr)/wid)**2)

def loss_function(args, vector, hist, func):
    prediction = func(np.arange(len(vector)), *args)
    where = np.argwhere(vector > prediction)
    
    energy = np.sum(np.gradient(prediction-vector)**2)
    sum_of_squares = np.sum(np.abs(vector - prediction))
    #soft_constraint_smaller = 0.5*np.sum(np.abs(vector[where]-prediction[where]))
    
    soft_constraint_equal_width = np.sum((args[::2]-np.mean(args[::2]))**2)
    loss = energy + sum_of_squares + soft_constraint_equal_width
    
    hist.append(loss)
    return loss
```

```python
def fit_gaussians(vector, loss_function, pre_filter = 15, initial_width = 50, initial_skew = 1, skewed = False):
    vector = vector.copy()
    minimum = np.min(vector)
    vector -= minimum
    vector = gaussian_filter(vector,pre_filter)
    x_maxs = maxima(vector, 0)
    y_maxs = vector[x_maxs]
    widths = np.ones(len(x_maxs))*initial_width
    hist = []
    gaussian = get_gaussian()
    if skewed:
        func = get_gaussian(centers = x_maxs, amplitudes = y_maxs, skewed=True)
        initial_skew = np.ones(len(x_maxs))*initial_skew
        initial_params = np.array(list(zip(widths,initial_skew )))
        res = minimize(lambda args: loss_function(args, vector, hist, func), x0 = initial_params)
        widths = np.array(res["x"])[::2]
        skews = np.array(res["x"])[1::2]
        gaussian_params = np.array(list(zip(x_maxs, y_maxs, widths, skews)))
        predictions = np.array([skewed_gaussian(np.arange(len(vector)),*p)+minimum for p in gaussian_params])
    else:
        func = get_gaussian(centers = x_maxs, amplitudes = y_maxs)
        print(widths)
        res = minimize(lambda args: loss_function(args, vector, hist, func), x0 = widths)
        gaussian_params = np.array(list(zip(x_maxs, y_maxs, res["x"])))
        predictions = np.array([gaussian(np.arange(len(vector)),*p)+minimum for p in gaussian_params])
    return predictions, gaussian_params, vector, hist
```
