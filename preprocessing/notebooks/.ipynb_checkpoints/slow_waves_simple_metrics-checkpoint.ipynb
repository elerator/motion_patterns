```python
def load_files_in_folder(folder):
    files = os.listdir(folder)
    files.sort()
    paths = [os.path.join(folder, f) for f in files]
    
    tensors = [np.load(f) for f in paths]
    return tensors, files
```

```python
def plot_animation(vectors, i_min = 0, n_frames = 10, norm=True):
    fig, ax = plt.subplots(1)
    p = ax.plot([])[0]
    def plot(i):
        plt.cla()
        for v in vectors:
            v = normalize(v[i_min+i])
            ax.plot(v)
    ani = matplotlib.animation.FuncAnimation(fig, plot, frames=n_frames)
    return ani
```

```python
levels = list(set(iso))
levels.sort(reverse=True)

for level in levels:
    cond = width[iso==level]
    y, x = np.histogram(cond, density = True, bins=100, range=[0,1500])
    x = 100/x
    plt.plot(x[:-1],gaussian_filter(y,1), label = str(level))
plt.legend()
```
